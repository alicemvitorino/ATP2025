```python
#b)

def comum(lista1, lista2):
    res = False
    for elem in lista1:
        if elem in lista2 and res == False:
            res = True
    return res
comum([1,2,3], [7,2,9])
comum([1,2,3], [7,5,9])
```

```python
#d)
hospital = [
(
"Unidade de Cardiologia",
[ (1, "Dr. João"), (2, "Dra. Maria")],
[(101, False), (102, True), (103, False), (104, False)]
),
(
"Unidade de Pediatria",
[ (3, "Dr. Pedro"), (4, "Dra. Clara") ],
[ (201, True), (202, True), (203, True)]
)
]

def existe_medico(hospital, unidade, nome_medico):
    res = False
    for elem in hospital:
        if elem[0] == unidade:
            for elem1 in elem[1]:
                if elem1[1] == nome_medico:
                    res = True
    return res
existe_medico(hospital, "Unidade de Cardiologia", "Dr. João")
```
